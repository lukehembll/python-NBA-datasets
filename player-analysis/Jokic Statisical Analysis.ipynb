## Calculating the overall percentage of shots, Shot Percentage = Made shots / Total shots 

```python
shots = df1['EVENT_TYPE'].value_counts()
```

```python
shotPercentage = shots['Made Shot']/[shots['Made Shot'] + shots['Missed Shot']]*100
print(shotPercentage)
```

## Calculating the percentage efficiency from each range

```python
df1['SHOT_ZONE_RANGE'].value_counts()
```

```python
df2 = df1.filter(['EVENT_TYPE','SHOT_ZONE_RANGE'])
```

```python
df2.head(10)
```

```python
df2['SHOT_ZONE_RANGE'].value_counts()
```

```python
df2['SHOT_ZONE_RANGE'].unique()
```

```python
df2.loc[df['EVENT_TYPE'] == 'Made Shot'].value_counts('SHOT_ZONE_RANGE').reindex(
    df2.SHOT_ZONE_RANGE.unique(), fill_value=0)
```

```python
fieldgoalsattempted = df2['SHOT_ZONE_RANGE'].value_counts().rename_axis('RANGE').reset_index(name='FGA')
print (fieldgoalsattempted)
```

```python
fieldgoalsmade = df2.loc[df['EVENT_TYPE'] == 'Made Shot'].value_counts('SHOT_ZONE_RANGE').reindex(
    df2.SHOT_ZONE_RANGE.unique(), fill_value=0).rename_axis('RANGE').reset_index(name='FGM')
print (fieldgoalsmade)
```

```python
fieldgoals = pd.merge(fieldgoalsmade, fieldgoalsattempted)
print(fieldgoals)
```

```python
fieldgoals['AVERAGE'] = (fieldgoals['FGM'] / fieldgoals['FGA'])*100
```

```python
avgField = fieldgoals.sort_values(by=['AVERAGE'], ascending=False)
print(avgField)
```

## Total Points (not including free throws)

```python
madeshots = df1.loc[df['EVENT_TYPE'] == 'Made Shot']
```

```python
madeshots.head()
```

```python
totalshots = madeshots['SHOT_TYPE'].value_counts().tolist()
print(totalshots)
```

```python
## 2 point field goal is worth 2 points, 3 point field goal is worth 3 points
```

```python
def points(lst):
    x=lst[0]
    y=lst[1]
    return (x*2) + (y*3)
```

```python
(points(totalshots))+379
```

```python
## Nikola Jokic made 379 free throws in the 2021-22 season
## each free throw is worth 1 point so we add free throws made to the total points.
```
